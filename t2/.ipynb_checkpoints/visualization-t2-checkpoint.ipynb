```python
import pandas as pd
import altair as alt
import datetime
```

```python
alt.data_transformers.disable_max_rows()
```

```python
data_file = '2014-2020.xlsx'
```

```python
data = pd.read_excel(data_file)
```

```python
data.head()
```

# Як змінювалась структура генерації електроенергії за роками?

```python
data[['time', 'date']] = data['Час/Дата'].str.split('-', expand = True)
data[['date', 'month', 'year']] = data['date'].str.split('.', expand = True)
data = data .filter(['time', 'date', 'month', 'year', 'AES', 'TEC', 'VDE', 'TES', 'GES', 'GAES_GEN',
       'CONSUMPTION'])
data.head()
```

```python
df1 = pd.melt(data, id_vars=['year'], value_vars=['AES', 'TEC', 'VDE', 'TES', 'GES', 'GAES_GEN'])
df1['Total'] = df1.groupby(['year', 'variable'])['value'].transform('sum')
df1 = df1.drop(['value'], axis=1)
df1.drop_duplicates(keep='first',inplace=True)
df1['year'] = pd.to_numeric(df1['year'])
df1['Total'] = pd.to_numeric(df1['Total'])
df1.head()
```

```python
alt.Chart(df1, title='Залежність структури електрогенерації в часі').transform_filter(alt.FieldRangePredicate(field='year', range=[2014, 2020])).mark_bar().encode(
    x = alt.X('year:O', title='Рік'),
    y = alt.Y('Total:Q', title="Згенерована електроенергія в МВт"),
    color = alt.Color('variable:N', title='Спосіб видобування', scale=alt.Scale(scheme='yelloworangered'))
).properties(
    width = 600, height = 500, background = '#F9F9F9', padding = 25
).configure_title(
    fontSize=32,
    font='Areal',
    offset=30,
).configure_axis(
    offset=12,
    labelFontSize=11,
    titleFontSize=11,
    titlePadding=17
)
```

Як на мене, є сенс використовувати bar-chart для цього завдання, оскільки він дозволяє показати загальну  кількість електроенергії згенерованої за рік, а різні кольори диференціюють типи електровидобуття.

# Як залежить споживання електроенергії від дня року та години доби

```python
df2 = data.drop(columns=['year', 'AES', 'TEC', 'VDE', 'TES', 'GES', 'GAES_GEN'])
df2['day'] = df2[['month', 'date']].agg('-'.join, axis=1)
df2 = df2.drop(columns=['date', 'month'])
df2['Total'] = df2.groupby(['time', 'day'])['CONSUMPTION'].transform('mean')
df2 = df2.drop(columns=['CONSUMPTION'])
df2.drop_duplicates(keep='first',inplace=True)
df2['time'] = pd.to_numeric(df2['time'])
df2['Total'] = pd.to_numeric(df2['Total'])
df2.head()
```

```python
alt.Chart(df2, title='Залежність споживання електроенергії від дня року та години доби').mark_rect().encode(
    alt.Y('time:O',  title='Година доби', sort=alt.EncodingSortField(field='time', order='descending')),
    alt.X('day:O', title='День в році', sort=alt.EncodingSortField(field='day', order='ascending')),
    color = alt.Color('Total:Q', title='Спожита енергія в МВт', scale=alt.Scale(scheme='redyellowgreen', reverse=True))
).properties(
    width = 700, height = 500, background = '#F9F9F9', padding = 25
).configure_title(
    fontSize=28,
    font='Areal',
    offset=30,
).configure_axis(
    offset=12,
    labelFontSize=11,
    titleFontSize=11,
    titlePadding=17
)
```

Тут я використовую heat-map. Кольором я кодую статистику по кількості спожитої електроенергії в певну годину певного дня

# Як змінюється генерація електроенергії з різних джерел впродовж доби

```python
df3 = pd.melt(data, id_vars=['time'], value_vars=['AES', 'TEC', 'VDE', 'TES', 'GES', 'GAES_GEN'])
df3['Total'] = df3.groupby(['time', 'variable'])['value'].transform('sum')
df3 = df3.drop(['value'], axis=1)
df3.drop_duplicates(keep='first',inplace=True)
df3['Total'] = pd.to_numeric(df3['Total'])
df3['time'] = pd.to_numeric(df3['time'])
df3.head()
```

```python
alt.Chart(df3, title='Залежність активності різних типів електрогенерації від години дня').transform_filter(alt.FieldRangePredicate(field='time', range=[0, 24])).mark_bar().encode(
    x = alt.X('time:N', title='Година', sort=alt.EncodingSortField(field='time', order='ascending')),
    y = alt.Y('Total:Q', title="Згенерована електроенергія в МВт", sort=alt.EncodingSortField(field='time', order='ascending')),
    color = alt.Color('variable:N', title='Спосіб видобування', scale=alt.Scale(scheme='yelloworangered'))
).properties(
    width = 700, height = 500, background = '#F9F9F9', padding = 25
).configure_title(
    fontSize=28,
    font='Areal',
    offset=30,
).configure_axis(
    offset=12,
    labelFontSize=11,
    titleFontSize=11,
    titlePadding=17
)
```

# Як змінюється споживання електроенергії впродовж доби у розрізі місяців року та пір року

```python
def label_season(month):
#     month = int(day.split('-')[0])
    if month in [12, 1, 2]:
        return 'winter'
    elif 3 <= month <= 5:
        return 'spring'
    elif 6 <= month <= 8:
        return 'summer'
    else:
        return 'autumn'
```

```python
df4 = data.drop(columns=['year', 'AES', 'TEC', 'VDE', 'TES', 'GES', 'GAES_GEN', 'date'])
df4['Mean'] = df4.groupby(['time', 'month'])['CONSUMPTION'].transform('mean')
df4 = df4.drop(columns=['CONSUMPTION'])
df4.drop_duplicates(keep='first',inplace=True)
for col in df4.columns:
    df4[col] = pd.to_numeric(df4[col])
df4['season'] = df4.apply (lambda row: label_season(row['month']), axis=1)
df4 = df4.sort_values(by=['month'])
df4['month'] = df4['month'].apply(lambda x: datetime.datetime.strptime(str(x), "%m").strftime("%b"))

df4
# len(data)
```

```python
alt.Chart(df4, title='Залежність споживання електроенергії від години в розрізі місяців та пір року').mark_rect().encode(
    x = alt.X('month:N', title='', sort=['Dec', 'Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov']),
    y = alt.Y('time:N', title='Година'),
    column = alt.Column('season:N', sort=["spring", "summer", "autumn", "winter"]),
    color = alt.Color('Mean', title='Спожита електроенергія в МВт', 
    scale = alt.Scale(scheme = 'redyellowblue', reverse=True))
).configure_title(
    fontSize=28,
    font='Areal',
    offset=30,
).resolve_scale(x='independent').properties(width = 180, height = 500, background = '#F9F9F9', padding = 25)
```
